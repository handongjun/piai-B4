```python
def show_function(students_list):
    students_list.sort(key=lambda e: e[4], reverse=True)#학점을 기준으로 내림차순으로 정렬
    print("Student", '\t', "   Name", "      Midterm", " Final", " Average"," Grade")
    print("--------------------------------------------------------------")
    for student in students_list:
        print(student[0], '\t',student[1], '\t', student[2], '\t', student[3], '\t', student[4], '\t ',student[5])


def search_function():
    stid_search = input("Student ID:")
    found_std = None
    
    # students_list 내에서 stid_search와 일치하는 학생 정보 찾기
    for student in students_list:
        if student[0] == str(stid_search):
            found_std = student
            break
            
    if found_std:
        print("Student", '\t', "   Name", "      Midterm", " Final", " Average"," Grade")
        print("--------------------------------------------------------------")
        print(found_std[0], '\t', found_std[1], '\t', found_std[2], '\t', found_std[3], '\t', found_std[4], '\t ', found_std[5])
    else:
        print("NO SUCH PERSON")

def changescore_function():
    stid_ch = input("Student ID:")
    found_std_ch = None
    
    for student in students_list:
        if student[0] == str(stid_ch):
            found_std_ch = student
            break

    if found_std_ch:
        mid_fin = input("Mid/Final :").lower()
        if mid_fin == "mid":
            new_mid = input("Input new score:")
            if int(new_mid) not in list(range(0,101)):
                print("Put 0~100 score")
            else:  
                print("Student", '\t', "   Name", "      Midterm", " Final", " Average"," Grade")
                print("--------------------------------------------------------------")
                print(found_std_ch[0], '\t', found_std_ch[1], '\t', found_std_ch[2], '\t', found_std_ch[3], '\t', found_std_ch[4], '\t ', found_std_ch[5])
                print("Score changed.")
                found_std_ch[2] = new_mid
                found_std_ch[4] = (int(found_std_ch[2]) + int(found_std_ch[3])) / 2
                if found_std_ch[4] >= 90:
                    found_std_ch[5] = "A"
                elif found_std_ch[4] >= 80:
                    found_std_ch[5] = "B"
                elif found_std_ch[4] >= 70:
                    found_std_ch[5] = "C"   
                elif found_std_ch[4] >= 60:
                    found_std_ch[5] = "D" 
                else:
                    found_std_ch[5] = "F"
                print(found_std_ch[0], '\t', found_std_ch[1], '\t', found_std_ch[2], '\t', found_std_ch[3], '\t', found_std_ch[4], '\t ', found_std_ch[5])

        elif mid_fin == "final":
            new_final = input("Input new score:")
            if int(new_final) not in list(range(0,101)):
                print("Put 0~100 score")
            else:
                print("Student", '\t', "   Name", "      Midterm", " Final", " Average"," Grade")
                print("--------------------------------------------------------------")
                print(found_std_ch[0], '\t', found_std_ch[1], '\t', found_std_ch[2], '\t', found_std_ch[3], '\t', found_std_ch[4], '\t ', found_std_ch[5])
                print("Score changed.")
                found_std_ch[3] = new_final
                found_std_ch[4] = (int(found_std_ch[2]) + int(found_std_ch[3])) / 2
                if found_std_ch[4] >= 90:
                    found_std_ch[5] = "A"
                elif found_std_ch[4] >= 80:
                    found_std_ch[5] = "B"
                elif found_std_ch[4] >= 70:
                    found_std_ch[5] = "C"   
                elif found_std_ch[4] >= 60:
                    found_std_ch[5] = "D" 
                else:
                    found_std_ch[5] = "F"
                print(found_std_ch[0], '\t', found_std_ch[1], '\t', found_std_ch[2], '\t', found_std_ch[3], '\t', found_std_ch[4], '\t ', found_std_ch[5])
        else:
            pass

    else:
        print("NO SUCH PERSON")



def searchgrade_function():
    grade_search = input("Grade to search:").upper()
    found_students = []
    for student in students_list:
        if student[5] == grade_search:
            if student not in found_students:  
                found_students.append(student)
                
    if grade_search not in ["A", "B", "C", "D", "F"]:
        print("Not Grade")

    elif found_students:
        print("Student", '\t', "  Name", "      Midterm", " Final", " Average"," Grade")
        print("--------------------------------------------------------------")
        for students in found_students:
            print(students[0], '\t', students[1], '\t', students[2], '\t', students[3], '\t', students[4], '\t ', students[5])
            
    else:
        print("NO RESULTS.")



def add_function():
    add_stid = input("Student ID:")
    found_std_add = None
    
    # students_list 내에서 add_stid와 일치하는 학생 정보 찾기
    for student in students_list:
        if student[0] == add_stid:
            found_std_add = student
            break
            
    if found_std_add:
        print("ALREADY EXISTS.")

    else:
        add_name = input("NAME:")
        add_mid = input("Midterm Score:")
        if int(add_mid) not in list(range(0,101)):
            print("Put 0~100 score")
        else:
            add_final = input("Final Score:")
            if int(add_final) not in list(range(0,101)):
                print("Put 0~100 score")
            else:
                add_avg = (int(add_mid) + int(add_final)) / 2
                if add_avg >= 90:
                    add_grade = "A"
                elif add_avg >= 80:
                    add_grade = "B"
                elif add_avg >= 70:
                    add_grade = "C"   
                elif add_avg >= 60:
                    add_grade = "D" 
                else:
                    add_grade = "F"
                students_list.append([add_stid, add_name, add_mid, add_final, add_avg, add_grade])
                print("Student added.")

def remove_function():
    if students_list == []:
        print("List is empty")

    else:
        remove_stid = input("Student ID:")
        found_std_remove = None
    
        # students_list 내에서 remove_stid와 일치하는 학생 정보 찾기
        for student in students_list:
            if student[0] == remove_stid:
                found_std_remove = student
                break

        if found_std_remove:
            students_list.remove(found_std_remove)
            print("Student removed.")

        elif students_list == []:
            print("List is empty")
            
        else:
            print("NO SUCH PERSON.")

def quit_function():
    save = input("save data?[yes/no]").lower()
    if save == "yes":
        file_name = input("File name:")
        students_list.sort(key=lambda e: e[4], reverse=True)#학점을 기준으로 내림차순으로 정렬
        with open(file_name, "w") as f:
            for student in students_list:
                f.write("\t".join(str(item) for item in student))
                f.write("\n")
    

    elif save == "no":
        print("quit")

    else:
        pass




with open("students.txt", 'r') as std:
    students_list = []
    for l in std:
        stid, name, mid, final = l.split("\t")
        final = int(final)
        #평균 및 학점 생성
        avg = (int(mid) + int(final)) / 2
        
        if avg >= 90:
            grade = "A"
        elif avg >= 80:
            grade = "B"
        elif avg >= 70:
            grade = "C"   
        elif avg >= 60:
            grade = "D" 
        else:
            grade = "F"

        students_list.append([stid, name, mid, final, avg, grade])

students_list.sort(key=lambda e: e[4], reverse=True)#학점을 기준으로 내림차순으로 정렬
print("Student", '\t', "   Name", "      Midterm", " Final", " Average"," Grade")
print("--------------------------------------------------------------")
for student in students_list:
    print(student[0], '\t',student[1], '\t', student[2], '\t', student[3], '\t', student[4], '\t ',student[5])


while True:
    command = input("# ")
    if command.lower() == "show":
        show_function(students_list)
    elif command.lower() == "search":
        search_function()
    elif command.lower() == "changescore":
        changescore_function()
    elif command.lower() == "searchgrade":
        searchgrade_function()
    elif command.lower() == "add":
        add_function()
    elif command.lower() == "remove":
        remove_function()    
    elif command.lower() == "quit":
        quit_function()
        break
    else:
        print("wrong input!")
```
